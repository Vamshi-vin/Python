# 🎯 Python Functions Practice

Functions are reusable blocks of code that perform specific tasks. This notebook covers function definition, parameters, return values, scope, decorators, and advanced patterns.

## 📘 Function Fundamentals

- **Definition**: `def function_name(parameters):` block
- **Parameters**: Input values (positional, keyword, default, *args, **kwargs)
- **Return values**: `return` statement; functions return `None` by default
- **Scope**: Local vs global variables; `global` and `nonlocal` keywords
- **Docstrings**: Documentation strings using triple quotes

```python
# basic function definition and call
def greet(name):
    return f'Hello, {name}!'

print(greet('Alice'))

# function with multiple parameters
def add(a, b):
    return a + b

print(add(3, 5))

# function with no return statement (returns None)
def say_goodbye():
    print('Goodbye!')

result = say_goodbye()
print('result:', result)
```

```python
# positional vs keyword arguments
def describe(name, age):
    return f'{name} is {age} years old'

# positional arguments
print(describe('Bob', 30))

# keyword arguments
print(describe(age=25, name='Charlie'))

# mix positional and keyword
print(describe('Diana', age=28))
```

```python
# default parameter values
def power(base, exponent=2):
    return base ** exponent

print(power(3))  # uses default exponent=2
print(power(3, 3))  # overrides default

# defaults must come after non-defaults
def greet_v2(greeting='Hello', name='World'):
    return f'{greeting}, {name}!'

print(greet_v2())
print(greet_v2(name='Python'))
print(greet_v2('Hi', 'There'))
```

```python
# *args - variable number of positional arguments (packed as tuple)
def sum_all(*args):
    print('args:', args, 'type:', type(args))
    return sum(args)

print(sum_all(1, 2, 3, 4, 5))

# **kwargs - variable keyword arguments (packed as dict)
def print_info(**kwargs):
    print('kwargs:', kwargs)
    for key, value in kwargs.items():
        print(f'  {key}: {value}')

print_info(name='Alice', age=30, city='NYC')

# combining normal, *args, and **kwargs
def flexible(a, b, *args, **kwargs):
    print(f'a={a}, b={b}')
    print(f'args={args}, kwargs={kwargs}')

flexible(1, 2, 3, 4, x=10, y=20)
flexible(5, 6, x=100)
```

```python
# unpacking arguments when calling
def multiply(a, b, c):
    return a * b * c

numbers = [2, 3, 4]
print('unpacking list with *:', multiply(*numbers))

# unpacking dict with **
def introduce(name, age, city):
    return f'{name}, {age}, {city}'

info = {'name': 'Bob', 'age': 25, 'city': 'LA'}
print('unpacking dict with **:', introduce(**info))
```

```python
# docstrings and help
def calculate_area(radius):
    """Calculate the area of a circle given its radius.
    
    Args:
        radius (float): The radius of the circle.
    
    Returns:
        float: The area of the circle.
    """
    return 3.14159 * radius ** 2

help(calculate_area)
print(calculate_area.__doc__)
```

```python
# returning multiple values
def get_min_max(lst):
    return min(lst), max(lst)  # returns a tuple

values = [3, 1, 4, 1, 5]
mn, mx = get_min_max(values)
print(f'min: {mn}, max: {mx}')

# unpacking return value
result = get_min_max(values)
print(f'result: {result}, type: {type(result)}')
```

```python
# scope: local vs global variables
global_var = 'I am global'

def scope_demo():
    local_var = 'I am local'
    print('inside function:')
    print('  local_var:', local_var)
    print('  global_var:', global_var)

scope_demo()
print('outside function:')
print('  global_var:', global_var)
# print('  local_var:', local_var)  # NameError if uncommented
```

```python
# modifying global variable
counter = 0

def increment():
    global counter  # declare global to modify
    counter += 1
    return counter

print(increment())
print(increment())
print('counter:', counter)
```

```python
#Type hints are annotations that indicate what data types your functions expect and return. 
# They're optional but highly recommended.
#  type hints (annotations)
def greet_typed(name: str) -> str:
    return f'Hello, {name}!'

print(greet_typed('Python'))

def add_typed(a: int, b: int) -> int:
    return a + b

print(add_typed(3, 7))

# hints don't enforce at runtime
print(add_typed('hello', 'world'))  # still works (concatenation)
```

```python
# lambda functions - anonymous, single-line functions
square = lambda x: x ** 2
print('lambda square(5):', square(5))

# lambda with multiple arguments
add_lambda = lambda x, y: x + y
print('lambda add(3, 4):', add_lambda(3, 4))

# lambda commonly used with map, filter, sorted
nums = [1, 2, 3, 4, 5]
squared = list(map(lambda x: x**2, nums))
print('map with lambda:', squared)

filtered = list(filter(lambda x: x > 2, nums))
print('filter with lambda:', filtered)
```

```python
# nested functions
def outer(x):
    def inner(y):
        return x + y
    return inner
#A closure is when an inner function "remembers" variables from its outer scope, 
# even after the outer function has finished executing.
add_5 = outer(5)  # returns the inner function
print('using nested function:', add_5(3))

# closure: inner function remembers outer's variables
def make_multiplier(n):
    def multiplier(x):
        return x * n  # remembers n from outer scope
    return multiplier

times_3 = make_multiplier(3)
print('closure - times_3(4):', times_3(4))
```

```python
# functions as first-class objects (can be passed around)
def apply_operation(a, b, operation):
    return operation(a, b)

def multiply(x, y):
    return x * y

print(apply_operation(3, 4, multiply))
print(apply_operation(3, 4, lambda x, y: x + y))

# storing functions in data structures
operations = {
    'add': lambda x, y: x + y,
    'sub': lambda x, y: x - y,
    'mul': lambda x, y: x * y,
}

print(operations['add'](5, 3))
print(operations['mul'](5, 3))
```

```python
# decorators - functions that modify other functions
def my_decorator(func):
    def wrapper():
        print('before calling function')
        func()
        print('after calling function')
    return wrapper

@my_decorator
def say_hello():
    print('Hello!')

say_hello()  # decorator wraps the function

# equivalent to: say_hello = my_decorator(say_hello)
```

```python
# decorator with arguments
def repeat(times):
    def decorator(func):
        def wrapper(*args, **kwargs):
            for _ in range(times):
                func(*args, **kwargs)
        return wrapper
    return decorator

@repeat(3)
def greet(name):
    print(f'Hello, {name}!')

greet('World')
```

## 🌍 Real-World Decorator Examples

Decorators are widely used in production code for logging, authentication, caching, timing, and validation. Here are practical examples you'll encounter in real applications.

```python
# 1. Timing decorator - measure function execution time
import time

def timer(func):
    def wrapper(*args, **kwargs):
        start = time.time()
        result = func(*args, **kwargs)
        end = time.time()
        print(f'{func.__name__} took {end - start:.4f} seconds')
        return result
    return wrapper

@timer
def slow_function():
    time.sleep(0.5)
    return 'done'

slow_function()
```

```python
# 2. Logging decorator - track function calls
def log_calls(func):
    def wrapper(*args, **kwargs):
        print(f'Calling {func.__name__} with args={args}, kwargs={kwargs}')
        result = func(*args, **kwargs)
        print(f'{func.__name__} returned {result}')
        return result
    return wrapper

@log_calls
def add_numbers(a, b):
    return a + b

add_numbers(5, 3)
```

```python
# 3. Caching decorator - memoization for expensive computations
def memoize(func):
    cache = {}
    def wrapper(n):
        if n not in cache:
            print(f'Computing {func.__name__}({n})')
            cache[n] = func(n)
        else:
            print(f'Returning cached result for {func.__name__}({n})')
        return cache[n]
    return wrapper

@memoize
def fibonacci(n):
    if n <= 1:
        return n
    return fibonacci(n-1) + fibonacci(n-2)

print(fibonacci(5))
print(fibonacci(5))  # uses cache, no recomputation
```

```python
# 4. Validation decorator - check preconditions
def validate_positive(func):
    def wrapper(*args, **kwargs):
        for arg in args:
            if isinstance(arg, (int, float)) and arg < 0:
                raise ValueError(f'Expected positive number, got {arg}')
        return func(*args, **kwargs)
    return wrapper

@validate_positive
def calculate_square_root(x):
    return x ** 0.5

print(calculate_square_root(16))
# calculate_square_root(-5)  # would raise ValueError
```

```python
# 5. Authentication decorator - check permissions before executing
def require_auth(user_role_required):
    def decorator(func):
        def wrapper(user_role, *args, **kwargs):
            if user_role != user_role_required:
                raise PermissionError(f"Requires {user_role_required} role, but got {user_role}")
            return func(*args, **kwargs)
        return wrapper
    return decorator

@require_auth('admin')
def delete_user(user_id):
    return f'User {user_id} deleted'

print(delete_user('admin', 123))
# delete_user('user', 123)  # would raise PermissionError
```

```python
# 6. Error handling decorator - catch and handle exceptions
def handle_errors(func):
    def wrapper(*args, **kwargs):
        try:
            return func(*args, **kwargs)
        except ZeroDivisionError:
            print(f'Error: Cannot divide by zero in {func.__name__}')
            return None
        except Exception as e:
            print(f'Unexpected error in {func.__name__}: {e}')
            return None
    return wrapper

@handle_errors
def divide(a, b):
    return a / b

print(divide(10, 2))
print(divide(10, 0))  # handled gracefully
```

```python
# map() - apply function to each element
def square(x):
    return x ** 2

nums = [1, 2, 3, 4]
squared = map(square, nums)
print('map result:', list(squared))

# filter() - keep elements where function returns True
def is_even(x):
    return x % 2 == 0

filtered = filter(is_even, nums)
print('filter result:', list(filtered))

# reduce() - combine elements using function
from functools import reduce
def multiply(a, b):
    return a * b

product = reduce(multiply, [2, 3, 4])
print('reduce result:', product)  # (2*3)*4 = 24
```

## ✅ Function Best Practices

- Keep functions small and focused (single responsibility).
- Use meaningful names that describe what the function does.
- Write docstrings for all public functions.
- Use type hints for clarity.
- Avoid side effects when possible (pure functions).
- Return early to reduce nesting and improve readability.

```python
# good practice: early return to flatten logic
def classify_number(n):
    if n < 0:
        return 'negative'
    if n == 0:
        return 'zero'
    if n < 10:
        return 'single digit'
    return 'multi-digit'

print(classify_number(-5))
print(classify_number(0))
print(classify_number(7))
print(classify_number(42))
```

```python
# practical: data validation function
def validate_email(email: str) -> bool:
    """Check if email has basic valid format.
    
    Args:
        email: Email string to validate.
    
    Returns:
        True if valid, False otherwise.
    """
    if '@' not in email or '.' not in email:
        return False
    return email.count('@') == 1

print(validate_email('user@example.com'))
print(validate_email('invalid.email'))
print(validate_email('user@@example.com'))
```
