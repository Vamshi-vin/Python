# 📁 Python File Handling Practice

File handling is essential for reading, writing, and managing files. This notebook covers opening/closing files, reading/writing operations, context managers, file modes, and practical file operations.

## 📘 File Handling Fundamentals

- **File modes**: `r` (read), `w` (write), `a` (append), `x` (create), `b` (binary), `t` (text)
- **Reading**: `read()` (all), `readline()` (one line), `readlines()` (all as list)
- **Writing**: `write()` (string), `writelines()` (list of strings)
- **Context manager**: `with` statement for automatic file closing
- **File operations**: Path handling, directory creation, file deletion
- **Exception handling**: Catching `FileNotFoundError`, `IOError`, etc.

```python
# create a sample file for practice
with open('sample.txt', 'w') as f:
    f.write('Hello, World!\n')
    f.write('This is line 2\n')
    f.write('This is line 3\n')

print('File created successfully')
```

```python
# reading entire file at once
with open('sample.txt', 'r') as f:
    content = f.read()  # reads all content as one string
    print('Using read():')
    print(content)

# reading line by line
with open('sample.txt', 'r') as f:
    line = f.readline()  # reads one line
    print('Using readline():')
    print(line, end='')  # line already has \n

# reading all lines into a list
with open('sample.txt', 'r') as f:
    lines = f.readlines()  # reads all lines as list
    print('\nUsing readlines():')
    print(lines)
```

```python
# iterating over file line by line (memory-efficient)
with open('sample.txt', 'r') as f:
    for line_num, line in enumerate(f, start=1):
        print(f'Line {line_num}: {line}', end='')
```

```python
# writing to a file (overwrites existing content)
with open('output.txt', 'w') as f:
    f.write('First line\n')
    f.write('Second line\n')
    f.writelines(['Third line\n', 'Fourth line\n'])

print('File written successfully')

# verify content
with open('output.txt', 'r') as f:
    print(f.read())
```

```python
# appending to a file (adds to existing content)
with open('output.txt', 'a') as f:
    f.write('Appended line\n')

print('Content after appending:')
with open('output.txt', 'r') as f:
    print(f.read())
```

```python
# file modes
# 'r'  - read (default)
# 'w'  - write (truncates file)
# 'a'  - append
# 'x'  - create (fails if exists)
# 'b'  - binary (rb, wb, ab)
# 't'  - text (rt, wt, at) - default

# create a new file with 'x' mode
try:
    with open('newfile.txt', 'x') as f:
        f.write('New file content')
    print('New file created')
except FileExistsError:
    print('File already exists')
```

```python
# context manager (with statement) - closes file automatically
# this is the preferred way
with open('sample.txt', 'r') as f:
    data = f.read()
# file is closed automatically here, no need for f.close()

# without context manager (not recommended)
f = open('sample.txt', 'r')
data = f.read()
f.close()  # must close manually
```

```python
# exception handling with files
try:
    with open('nonexistent.txt', 'r') as f:
        content = f.read()
except FileNotFoundError:
    print('Error: File not found')
except IOError as e:
    print(f'IO Error: {e}')

# checking if file exists before opening
import os
if os.path.exists('sample.txt'):
    print('sample.txt exists')
else:
    print('sample.txt does not exist')
```

```python
# working with file paths
import os
from pathlib import Path

# using os.path
file_path = 'sample.txt'
print('File name:', os.path.basename(file_path))
print('Directory:', os.path.dirname(file_path))
print('Absolute path:', os.path.abspath(file_path))

# using pathlib (modern approach)
p = Path('sample.txt')
print('\nUsing pathlib:')
print('Name:', p.name)
print('Parent:', p.parent)
print('Absolute:', p.absolute())
```

```python
# directory operations
import os

# create a directory
if not os.path.exists('my_directory'):
    os.mkdir('my_directory')
    print('Directory created')

# create nested directories
os.makedirs('nested/dir/structure', exist_ok=True)
print('Nested directories created')

# list files in directory
print('\nFiles in current directory:')
for file in os.listdir('.'):
    if file.endswith('.txt'):
        print(f'  {file}')
```

```python
# working with CSV files
import csv

# write CSV file
data = [
    ['Name', 'Age', 'City'],
    ['Alice', 30, 'NYC'],
    ['Bob', 25, 'LA'],
    ['Charlie', 35, 'Chicago']
]

with open('data.csv', 'w', newline='') as f:
    writer = csv.writer(f)
    writer.writerows(data)
print('CSV file written')

# read CSV file
with open('data.csv', 'r') as f:
    reader = csv.reader(f)
    for row in reader:
        print(row)
```

```python
# working with JSON files
import json

# write JSON file
person = {
    'name': 'Alice',
    'age': 30,
    'skills': ['Python', 'SQL', 'JavaScript'],
    'address': {
        'city': 'NYC',
        'country': 'USA'
    }
}

with open('person.json', 'w') as f:
    json.dump(person, f, indent=2)
print('JSON file written')

# read JSON file
with open('person.json', 'r') as f:
    loaded_person = json.load(f)
    print('Loaded data:')
    print(loaded_person)
```

```python
# reading file in binary mode
# useful for images, executables, etc.
with open('sample.txt', 'rb') as f:
    binary_data = f.read()
    print('Binary data:', binary_data)

# writing binary data
with open('binary_output.bin', 'wb') as f:
    f.write(b'\x00\x01\x02\x03')
print('Binary file written')
```

```python
# file position and seeking
with open('sample.txt', 'r') as f:
    # read first 5 characters
    first_five = f.read(5)
    print('First 5 chars:', first_five)

    # current position
    print('Current position:', f.tell())

    # seek to beginning
    f.seek(0)
    all_content = f.read()
    print('After seeking to 0:', all_content)
```

```python
# copying files
import shutil

shutil.copy('sample.txt', 'sample_backup.txt')
print('File copied')

# moving/renaming files
shutil.move('sample_backup.txt', 'backup/sample_backup.txt')
print('File moved')
```

```python
# deleting files
import os

if os.path.exists('newfile.txt'):
    os.remove('newfile.txt')
    print('File deleted')

# remove directory
import shutil
if os.path.exists('my_directory'):
    shutil.rmtree('my_directory')
    print('Directory removed')
```

## ✅ File Handling Best Practices

- Always use `with` statement for file operations (automatic closing).
- Handle exceptions for file not found and IO errors.
- Use `pathlib.Path` for modern path handling.
- Check file existence before reading or deleting.
- Use appropriate file modes (`r`, `w`, `a`).
- Use `csv` and `json` modules for structured data.
- Close files explicitly only if not using `with` statement.
- Use binary mode (`rb`, `wb`) for non-text files.

```python
# practical: read file, process, write results
def process_file(input_file, output_file):
    """Read file, convert lines to uppercase, write to output."""
    try:
        with open(input_file, 'r') as infile, open(output_file, 'w') as outfile:
            for line in infile:
                outfile.write(line.upper())
        print(f'Processed {input_file} -> {output_file}')
    except FileNotFoundError as e:
        print(f'Error: {e}')

process_file('sample.txt', 'uppercase_output.txt')
```

---
**Key takeaways:** Always use `with` for file operations, handle exceptions, use appropriate file modes, and leverage `csv`/`json` modules for structured data.
