# Python `lambda` Functions — Detailed Guide

This notebook explains `lambda` (anonymous) functions in Python with practical examples, common patterns, gotchas, and real-world use cases (map/filter/sorted/pandas).

## 1. Syntax and basics

```python
# basic lambda that adds two numbers
add = lambda x, y: x + y
print(add(2, 3))  # 5
# equivalent def form
def add_def(x, y):
    return x + y

print(add_def(2,3))
```

Notes:
- A `lambda` is a single expression; it implicitly returns that expression's value.
- Lambdas are syntactic sugar for small, throwaway functions; use `def` for complex logic.

## 2. Parameter variants: defaults, *args/**kwargs

```python
# default argument
f = lambda x, y=10: x * y
print(f(3), f(3,2))
# *args (pack) and **kwargs (pack) work in lambdas too
g = lambda *a, **kw: (a, kw)
print(g(1,2, name='a'))
```

## 3. Common functional patterns: map, filter, reduce, sorted key

```python
nums = [1,2,3,4,5]
# map: apply function to each element
print('map squares:', list(map(lambda x: x*x, nums)))
# filter: keep elements where predicate is True
print('even:', list(filter(lambda x: x%2==0, nums)))
# sorted with key
words = ['apple','Banana','cherry','date']
print('sorted case-insensitive:', sorted(words, key=lambda w: w.lower()))
# reduce (from functools) for cumulative operations
from functools import reduce
print('sum via reduce:', reduce(lambda a,b: a+b, nums))
```

## 4. Closures and lambdas capturing variables (late binding gotcha)

```python
# late binding example: creating list of lambdas in loop
funcs = []
for i in range(3):
    funcs.append(lambda: i)
print([f() for f in funcs], '(captured incorrectly - all return 2)')

# fixed version using default argument to capture value at definition time
funcs_fixed = []
for i in range(3):
    funcs_fixed.append(lambda x=i: x)
print([f() for f in funcs_fixed], '(captured correctly)')
```

## 5. Readability and when not to use lambda

- Prefer `def` when logic is multi-line, needs a docstring, or will be reused.
- Lambdas are great for short callbacks, small key functions, or inline simple transforms.
- Avoid nested complex lambdas; they reduce readability.

## 6. Real-world examples — pandas, sorting, event handlers

```python
import pandas as pd
df = pd.DataFrame({'name':['alice','bob','carol'], 'age':[25,30,35], 'score':[88,92,79]})
# inline transform with lambda in assign (though vectorized ops preferred)
print(df.assign(score_norm=lambda d: (d.score - d.score.mean()) / d.score.std()))
# groupby + apply with lambda (small functions)
print(df.groupby(df.age // 10).apply(lambda d: d.score.mean()))
# use lambda as key in sorted to sort by last character
names = ['anne','zoe','mark']
print(sorted(names, key=lambda n: n[-1]))
```

## 7. Higher-order functions & returning lambdas

```python
def make_multiplier(m):
    return lambda x: x * m
double = make_multiplier(2)
print(double(5))  # 10
# equivalent with def to show same pattern
def make_multiplier_def(m):
    def inner(x):
        return x * m
    return inner

print(make_multiplier_def(3)(5))
```

## 8. Lambda alternatives and tools (operator, itemgetter, attrgetter)

```python
from operator import itemgetter, attrgetter, methodcaller
pairs = [(1,'a'), (2,'b'), (3,'c')]
print('sorted by second item with itemgetter:', sorted(pairs, key=itemgetter(1)))
# itemgetter/attrgetter are often faster and clearer than lambda for simple attribute/item access
```
