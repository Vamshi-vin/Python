# Variables, Operators, Keywords, and Data Types — Practice

Concise reference and runnable examples for Python basics: variable assignment, operators, built-in keywords, and core data types. Useful for interviews and quick refreshes.

## Variables and assignment

```python
# dynamic typing: variables can be rebound to different types
x = 10
print(type(x), x)
x = 'hello'
print(type(x), x)
# multiple assignment and swapping
a, b = 1, 2
a, b = b, a  # swap without temp
print('a=', a, 'b=', b)
# chained assignment
c = d = 0
print('c,d =', c, d)
```

## Names, scope and `id()` (object identity)

```python
x = 256
y = 256
print('small-int caching: id(x)==id(y)?', id(x)==id(y))
x = 1000
y = 1000
print('large-int ids equal?', id(x)==id(y))
# mutable object identity
L = [1,2,3]
M = L
print('L is M?', L is M)
M.append(4)
print('L after M.append:', L)
```

## Operators — arithmetic, comparison, logical, bitwise, assignment

```python
# arithmetic
print('+, -, *, /, //, %, **')
print('5/2 =', 5/2, '5//2 =', 5//2, '5%2 =', 5%2, '5**2 =', 5**2)
# comparison
print('==, !=, <, <=, >, >= ->', 1 == 1, 1 != 2, 2 > 1)
# logical
print('and, or, not ->', True and False, True or False, not False)
# bitwise
print('&, |, ^, ~, <<, >> ->', 6 & 3, 6 | 3, 6 ^ 3, ~1, 1 << 2, 8 >> 2)
# augmented assignment
x = 1
x += 2
print('x after += 2 ->', x)
```

## Operator precedence (quick reminder)

```python
# parentheses control order; exponentiation has higher precedence than unary minus
print('-3**2 =', -3**2)
print('(-3)**2 =', (-3)**2)
print('2 + 3*4 =', 2 + 3*4)
```

## Python keywords (inspect programmatically)

```python
import keyword
print('Number of keywords:', len(keyword.kwlist))
print(keyword.kwlist)
```

## Core data types and short examples

```python
# numbers: int, float, complex
i = 42
f = 3.14
c = 1+2j
print(i, type(i), f, type(f), c, type(c))
# boolean and None
b = True
n = None
print(b, type(b), n, type(n))
# text types
s = 'hello'
print(s.upper(), s[1:4])
# bytes
by = b'abc'
print(by, type(by))
# sequences: list, tuple, range
lst = [1,2,3]
tpl = (1,2,3)
rng = range(5)
print(lst, tpl, list(rng))
# mappings & sets
d = {'a':1}
st = {1,2,3}
print(d, st)
```

## Mutability & common pitfalls

```python
# lists are mutable; tuples are not (shallow immutability)
a = [1,2]
b = a
b.append(3)
print('a after b.append ->', a)
t = (1, [2,3])  # tuple with mutable element
t[1].append(4)
print('tuple with mutated element ->', t)
```

## Type checking and conversion

```python
x = 10
print(isinstance(x, int))
print(type(x) is int)
# conversions
print(int(3.9), float('2.5'), str(123), list((1,2)))
```

## Naming conventions and best practices

- Use lowercase_with_underscores for variables and functions.
- UPPERCASE for constants (convention only).
- Avoid shadowing built-ins (e.g., don't name a variable `list`).
- Prefer descriptive names for readability.
