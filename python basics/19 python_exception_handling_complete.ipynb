## 1. What is an Exception?

```python
x = 10
y = 0

# This will raise ZeroDivisionError
result = x / y
print(result)
```

Exceptions stop program execution unless handled.

## 2. Basic try / except

```python
try:
    x = 10
    y = 0
    result = x / y
except ZeroDivisionError:
    print('Cannot divide by zero')
```

## 3. Catching Multiple Exceptions

```python
try:
    num = int('abc')
    value = 10 / num
except ValueError:
    print('Conversion failed')
except ZeroDivisionError:
    print('Division by zero')
```

## 4. Handling Multiple Exceptions Together

```python
try:
    num = int('abc')
except (ValueError, TypeError):
    print('Invalid conversion')
```

## 5. Using else Block

```python
try:
    num = int('5')
except ValueError:
    print('Conversion failed')
else:
    print('Conversion successful:', num)
```

## 6. Using finally Block

```python
try:
    f = open('sample.txt')
except FileNotFoundError:
    print('File not found')
finally:
    print('Execution completed')
```

## 7. Getting Exception Information

```python
try:
    num = int('abc')
except Exception as e:
    print('Error occurred:', e)
```

## 8. Raising Exceptions

```python
age = -5

if age < 0:
    raise ValueError('Age cannot be negative')
```

## 9. Custom Exceptions

```python
class InsufficientBalanceError(Exception):
    pass

def withdraw(balance, amount):
    if amount > balance:
        raise InsufficientBalanceError('Not enough balance')
    return balance - amount

withdraw(1000, 2000)
```

## 10. Exception Chaining 

Exception chaining means linking one exception to another so that Python shows the original cause of the error along with the new exception.

This is useful when you catch an exception and raise another exception, but still want to preserve the original error information.

```python
try:
    num = int('abc')
except ValueError as e:
    raise RuntimeError('Conversion failed') from e
```

## 11. Assertions

```python
x = 5
assert x > 0
print('Assertion passed')
```

## 12. Best Practices

- Catch specific exceptions instead of generic ones
- Avoid bare except blocks
- Use finally for cleanup operations
- Log exceptions in production systems
- Create custom exceptions for domain errors
