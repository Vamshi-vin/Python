```python
#explain object-oriented programming in python
#  Object-oriented programming (OOP) is a programming paradigm that uses "objects" to design applications and computer programs. 
# In Python, OOP allows you to create classes, which are blueprints for creating objects. 
# An object is an instance of a class, and it can have attributes (data) and methods (functions) that operate on the data.  
# Here is a simple example of a class in Python:
class Human:
    def __init__(self, name, age):
        self.name = name  # attribute
        self.age = age    # attribute

    def speak(self):  # method
        return f"{self.name} says hello!"
    
# Creating an instance of the Human class
my_human = Human("Vamshi", 38)
print(my_human.name)  # Output: Vamshi
print(my_human.age)   # Output: 38
print(my_human.speak()) # Output: Vamshi says hello!

# In this example, we defined a class called Human with an __init__ method that initializes the name and age attributes.
# We also defined a method called speak that returns a string when called.
# We then created an instance of the Human class called my_human and accessed its attributes and method.        
```

# OOP Concepts in Python

## 1. Encapsulation
Encapsulation is the concept of bundling data and the methods that operate on that data within a single unit, called a class.

It also restricts direct access to some of an object's components, which helps prevent accidental modification of data. In Python, this is commonly done using private attributes.

## 2. Inheritance
Inheritance allows a new class, called a child class, to inherit attributes and methods from an existing class, called a parent class.

This promotes code reusability. In Python, you can create a child class by passing the parent class as an argument.

## 3. Polymorphism
Polymorphism allows methods to behave differently depending on the object they are acting upon.

In Python, polymorphism is often achieved through method overriding, where a child class provides a specific implementation of a method already defined in its parent class.

## 4. Abstraction
Abstraction is the concept of hiding complex implementation details and showing only the essential features of an object.

In Python, abstraction can be achieved using abstract classes and abstract methods from the `abc` module.


```python
#encasulation example covering all the concepts of encapsulation in python 
# Encapsulation is one of the fundamental principles of object-oriented programming. It refers to the bundling of data (attributes) and methods (functions) that operate on the data into a single unit, or class.
# Encapsulation also involves restricting direct access to some of an object's components, which is a means of preventing accidental interference and misuse of the data. In Python, this is typically achieved using private attributes     

# In this example, we have a BankAccount class with a private attribute __balance.
# The deposit and withdraw methods allow us to modify the balance, while the get_balance method allows us to access the balance without directly accessing the private attribute.
# This encapsulation ensures that the balance can only be modified through the defined methods, preventing unauthorized
# access and maintaining the integrity of the data.

class BankAccount:
    def __init__(self, owner, balance=0):
        self.owner = owner  # public attribute
        self.__balance = balance  # private attribute

    def deposit(self, amount):
        if amount > 0:
            self.__balance += amount
            print(f"Deposited {amount}. New balance is {self.__balance}.")
        else:
            print("Deposit amount must be positive.")

    def withdraw(self, amount):
        if amount > 0 and amount <= self.__balance:
            self.__balance -= amount
            print(f"Withdrew {amount}. New balance is {self.__balance}.")
        else:
            print("Withdrawal amount must be positive and less than or equal to the balance.")

    def get_balance(self):
        return self.__balance   
    
# Creating an instance of BankAccount
account = BankAccount("Vamshi", 1000)
print(account.owner)  # Output: Vamshi
account.deposit(500)  # Output: Deposited 500. New balance is 1500.
account.withdraw(200) # Output: Withdrew 200. New balance is 1300.
#print(account.__balance)  # This will raise an AttributeError because __balance is a private attribute
print(account.get_balance()) # Output: 1300             


```

```python
# Inheritance is a fundamental concept in object-oriented programming that allows a new class (called a child class or subclass) to inherit attributes and methods from an existing class (called a parent class or superclass). This promotes code reusability and establishes a natural hierarchical relationship between classes.
# In Python, you can create a child class that inherits from a parent class by specifying the parent class in parentheses after the child class name. The child class can override methods from the parent class or add new methods and attributes to extend the functionality of the parent class.
# Here is an example of inheritance in Python:
# Parent class (superclass)
# Bank account example with inheritance

class BankAccount:
    def __init__(self, owner, balance=0):
        self.owner = owner
        self.__balance = balance

    def deposit(self, amount):
        if amount > 0:
            self.__balance += amount
            print(f"Deposited {amount}. New balance is {self.__balance}.")
        else:
            print("Deposit amount must be positive.")

    def withdraw(self, amount):
        if amount > 0 and amount <= self.__balance:
            self.__balance -= amount
            print(f"Withdrew {amount}. New balance is {self.__balance}.")
        else:
            print("Withdrawal amount must be positive and less than or equal to the balance.")

    def get_balance(self):
        return self.__balance
    
# Child class (subclass) that inherits from BankAccount
class SavingsAccount(BankAccount):
    def __init__(self, owner, balance=0, interest_rate=0.01):
        super().__init__(owner, balance)  # Call the parent class constructor
        self.interest_rate = interest_rate  # New attribute specific to SavingsAccount

    def apply_interest(self):
        interest = self.get_balance() * self.interest_rate
        self.deposit(interest)  # Use the deposit method from the parent class to add interest  
# Creating an instance of SavingsAccount
savings_account = SavingsAccount("Vamshi", 1000, 0.05)
print(savings_account.owner)  # Output: Vamshi
print(savings_account.get_balance())  # Output: 1000
savings_account.apply_interest()  # Output: Deposited 50.0. New balance
# is 1050.0
print(savings_account.get_balance())  # Output: 1050.0  
```

```python
#another example of inheritance with a different context
# Parent class (superclass)
class Vehicle:
    def __init__(self, make, model):
        self.make = make
        self.model = model

    def start_engine(self):
        return f"The {self.make} {self.model}'s engine has started."    
# Child class (subclass) that inherits from Vehicle
class Car(Vehicle):
    def __init__(self, make, model, num_doors):
        super().__init__(make, model)  # Call the parent class constructor
        self.num_doors = num_doors  # New attribute specific to Car
    def open_trunk(self):
        return f"The trunk of the {self.make} {self.model} is now open."
# Creating an instance of Car
my_car = Car("Honda", "City", 4)
print(my_car.make)  # Output: Toyota
print(my_car.model)  # Output: Camry
print(my_car.num_doors)  # Output: 4
print(my_car.start_engine())  # Output: The Toyota Camry's engine has started.
print(my_car.open_trunk())  # Output: The trunk of the Toyota Camry is
# now open. 
```

```python
#explain tricky concepts of inheritance in python with examples
# Inheritance in Python can sometimes lead to tricky situations, especially when dealing with multiple inheritance (where a class inherits from more than one parent class) or when overriding methods in child classes. Here
# are some examples to illustrate these concepts:
# Multiple Inheritance Example
class A:
    def method_a(self):
        return "Method A from class A"  
class B:
    def method_b(self):
        return "Method B from class B"
class C(A, B):
    def method_c(self):
        return "Method C from class C"
# Creating an instance of class C
c_instance = C()
print(c_instance.method_a())  # Output: Method A from class A
print(c_instance.method_b())  # Output: Method B from class B
print(c_instance.method_c())  # Output: Method C from class C       
```

```python
#interview question on inheritance in python
# Question: What is the method resolution order (MRO) in Python, and how does it affect multiple inheritance?
# Answer: The method resolution order (MRO) in Python is the order in which Python looks for a method in a hierarchy of classes when multiple inheritance is involved. It determines the order in which base classes are searched when executing a method. The MRO is determined using
# the C3 linearization algorithm, which ensures a consistent method resolution order while respecting the hierarchy of classes. 
# The MRO can be accessed using the __mro__ attribute or the mro() method of a class. When a method is called on an instance of a class that has multiple parent classes, 
# Python will search for the method in the order defined by the MRO, starting from the child class and moving up to the parent classes. This helps to avoid ambiguity and ensures that the correct method is executed when multiple classes have methods with the same name.  
# Example of MRO in multiple inheritance
class X:
    def method(self):
        return "Method from class X"
class Y:
    def method(self):
        return "Method from class Y"
class Z(X, Y):
    pass    
# Creating an instance of class Z
z_instance = Z()
print(z_instance.method())  # Output: Method from class X
print(Z.__mro__)  # Output: (<class '__main__.Z'>, <class '__main__.X'>, <class '__main__.Y'>, <class 'object'>
# In this example, class Z inherits from both class X and class Y. 
# When we call the method() on an instance of class Z, 
# Python looks for the method in the order defined by the MRO. 
# Since class X is listed first in the MRO, its method() is called.
```

```python
#explain polymorphism in python with examples
# Polymorphism is a fundamental concept in object-oriented programming that allows objects of different classes to be treated as instances of the same class through a common interface. In Python, polymorphism is achieved through method overriding and duck typing. 
# Method overriding occurs when a child class provides a specific implementation of a method that is already defined in its parent class. This allows the child class to modify or extend the behavior of the parent class method. 
# Duck typing is a concept in Python where the type of an object is determined by its behavior (methods and properties) rather than its class. This allows for greater flexibility and enables polymorphism without the need for a common base class. 
# Here are some examples to illustrate polymorphism in Python:
# Method Overriding Example
class Animal:
    def speak(self):
        return "Animal makes a sound"
class Dog(Animal):
    def speak(self):
        return "Dog barks"
class Cat(Animal):
    def speak(self):
        return "Cat meows"
# Creating instances of Dog and Cat
dog = Dog()
cat = Cat()
print(dog.speak())  # Output: Dog barks
print(cat.speak())  # Output: Cat meows
# In this example, both Dog and Cat classes override the speak() method of the Animal class. 
# When we call the speak() method on instances of Dog and Cat, the respective overridden methods are executed, demonstrating polymorphism.              
```

```python
# Duck Typing Example
class Bird:
    def fly(self):
        return "Bird is flying"
class Airplane:
    def fly(self):
        return "Airplane is flying"
def make_it_fly(flyable):
    return flyable.fly()
# Creating instances of Bird and Airplane
bird = Bird()
airplane = Airplane()
print(make_it_fly(bird))  # Output: Bird is flying
print(make_it_fly(airplane))  # Output: Airplane is flying      
# In this example, both Bird and Airplane classes have a fly() method.
# The make_it_fly() function takes any object that has a fly() method and calls it. 
# This demonstrates duck typing, as we can pass instances of different classes to the same function, and it will work as long as they have the required method, showcasing polymorphism without a common base class.
```

```python
#method overloading is not directly supported in Python, 
# but we can achieve similar functionality using default arguments or variable-length arguments.
# Method Overloading using Default Arguments
class Calculator:
    def add(self, a, b=0):
        return a + b
# Creating an instance of Calculator
calc = Calculator()
print(calc.add(5))    # Output: 5 (using default value for b)
print(calc.add(5, 10)) # Output: 15 (using provided value for b)
# Method Overloading using Variable-Length Arguments
class Calculator:
    def add(self, *args):
        return sum(args)
# Creating an instance of Calculator
calc = Calculator()
print(calc.add(5))          # Output: 5
print(calc.add(5, 10))      # Output: 15
print(calc.add(5, 10, 15))  # Output: 30
# In these examples, we have a Calculator class with an add() method that can handle different numbers of arguments. 
# The first example uses default arguments to allow for method overloading, while the second example uses variable-length arguments (*args) to achieve a similar effect, demonstrating how we can   simulate method overloading in Python.  
```

```python
#any other tricky concepts of polymorphism in python with examples
# One tricky concept of polymorphism in Python is the use of abstract base classes (ABCs) and the abc module. Abstract base classes allow you to define a common interface for a group of related classes, and they can enforce that certain methods are implemented in the child classes. This is particularly
# useful when you want to ensure that all subclasses of a certain class implement specific methods, even if they have different implementations.
# Here is an example of using abstract base classes in Python:
from abc import ABC, abstractmethod
# Abstract base class
class Shape(ABC):
    @abstractmethod
    def area(self):
        pass        
# Child class that inherits from Shape
class Circle(Shape):
    def __init__(self, radius):
        self.radius = radius

    def area(self):
        return 3.14 * self.radius ** 2
# Child class that inherits from Shape
class Rectangle(Shape):
    def __init__(self, width, height):
        self.width = width
        self.height = height

    def area(self):
        return self.width * self.height
# Creating instances of Circle and Rectangle
circle = Circle(5)
rectangle = Rectangle(4, 6)
print(circle.area())  # Output: 78.5
print(rectangle.area())  # Output: 24       
# In this example, we have an abstract base class called Shape with an abstract method area().
# The Circle and Rectangle classes inherit from Shape and provide their own implementations of the area() method
```

```python
# Another tricky concept of polymorphism in Python is the use of operator overloading. Operator overloading allows you to define how operators (like +, -, *, etc.) behave for instances of your classes. This can make your classes more intuitive and easier to use, but it can also lead to confusion if not implemented carefully.
# Here is an example of operator overloading in Python:
class Vector:               
    def __init__(self, x, y):
        self.x = x
        self.y = y

    def __add__(self, other):
        if isinstance(other, Vector):
            return Vector(self.x + other.x, self.y + other.y)
        return NotImplemented

    def __str__(self):
        return f"Vector({self.x}, {self.y})"
# Creating instances of Vector
vector1 = Vector(2, 3)
vector2 = Vector(4, 5)
# Adding two vectors using the overloaded + operator
result_vector = vector1 + vector2
print(result_vector)  # Output: Vector(6, 8)
# In this example, we have a Vector class that overloads the + operator by defining the __add__ method. This allows us to add two Vector instances together in a natural way, demonstrating polymorphism through operator overloading. However, it's important to ensure that the overloaded operators are implemented in a way that is intuitive and consistent with the expected behavior to avoid confusion for users of the class.
```

```python
#abstract base classes and interfaces in python with examples
# In Python, abstract base classes (ABCs) are defined using the abc module. An abstract base class is a class that cannot be instantiated and is meant to be subclassed. It can
#define abstract methods that must be implemented by any subclass. This allows you to create a common interface for a group of related classes, ensuring that they all implement certain methods
from abc import ABC, abstractmethod
# Abstract base class
class Animal(ABC):
    @abstractmethod
    def speak(self):
        pass    
# Child class that inherits from Animal
class Dog(Animal):
    def speak(self):
        return "Dog barks"
# Child class that inherits from Animal
class Cat(Animal):
    def speak(self):
        return "Cat meows"
# Creating instances of Dog and Cat
dog = Dog()
cat = Cat()
print(dog.speak())  # Output: Dog barks
print(cat.speak())  # Output: Cat meows
# In this example, we have an abstract base class called Animal with an abstract method speak().
# The Dog and Cat classes inherit from Animal and provide their own implementations of the speak() method. 
# This demonstrates how abstract base classes can be used to define a common interface for related classes, ensuring that they all implement the required methods.
```

```python
# In Python, interfaces are not explicitly defined as in some other programming languages, but you can achieve a similar effect using abstract base classes. An interface is essentially an abstract base class that only contains abstract methods, which must be implemented by any subclass.
# Here is an example of using an abstract base class to define an interface in Python:
from abc import ABC, abstractmethod
# Interface (abstract base class with only abstract methods)
class Shape(ABC):
    @abstractmethod
    def area(self):
        pass
    @abstractmethod
    def perimeter(self):
        pass    
# Child class that implements the Shape interface
class Circle(Shape):
    def __init__(self, radius):
        self.radius = radius

    def area(self):
        return 3.14 * self.radius ** 2

    def perimeter(self):
        return 2 * 3.14 * self.radius
# Child class that implements the Shape interface
class Rectangle(Shape):
    def __init__(self, width, height):
        self.width = width
        self.height = height

    def area(self):
        return self.width * self.height

    def perimeter(self):
        return 2 * (self.width + self.height)
# Creating instances of Circle and Rectangle
circle = Circle(5)
rectangle = Rectangle(4, 6)
print(circle.area())  # Output: 78.5            
print(circle.perimeter())  # Output: 31.400000000000002
print(rectangle.area())  # Output: 24
print(rectangle.perimeter())  # Output: 20
# In this example, we have an abstract base class called Shape that defines two abstract methods: area() and perimeter(). 
# The Circle and Rectangle classes implement the Shape interface by providing concrete implementations of these methods. This allows us to treat instances of Circle and Rectangle as instances of Shape, demonstrating polymorphism through the use of interfaces in Python.
```
