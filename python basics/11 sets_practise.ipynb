# 📗 Python Sets Practice

A set is an unordered collection of unique, immutable elements. This notebook demonstrates creation, common operations, and set algebra (union, intersection, difference).

**Properties of sets:**

- **Unordered**: No guaranteed element order.
- **Unique elements**: Duplicates are removed.
- **Mutable (container)**: You can add/remove elements, but elements themselves must be hashable (immutable).
- **No indexing/slicing**: Use iteration instead.

**Common methods & operations**

- `add(elem)`, `update(iterable)`
- `remove(elem)` (raises `KeyError`), `discard(elem)` (no error), `pop()`
- `clear()`, `copy()`
- Set algebra: `union(|)`, `intersection(&)`, `difference(-)`, `symmetric_difference(^)`
- Tests: `issubset()`, `issuperset()`

```python
# creating sets
s1 = {1, 2, 3, 4}
s2 = set([3, 4, 5, 6])
s3 = set(range(1, 6))
print(s1, s2, s3)

# duplicates removed automatically
s_dup = {1, 2, 2, 3, 3}
print(s_dup)
```

```python
# membership and iteration (no indexing)
s = {10, 20, 30}
print(20 in s)
for x in s:
    print(x)

# adding and removing
s.add(40)
s.update([50, 60])
print(s)
s.discard(20)
print(s)
# s.remove(99)  # would raise KeyError if uncommented
print(s.pop())  # removes and returns an arbitrary element
print(s)
s.clear()
print(s)
```

```python
# set algebra
a = {1, 2, 3, 4}
b = {3, 4, 5, 6}
print('a | b ->', a | b)
print('a.union(b) ->', a.union(b))
print('a & b ->', a & b)
print('a.intersection(b) ->', a.intersection(b))
print('a - b ->', a - b)
print('a.symmetric_difference(b) ->', a ^ b)

# subset / superset
print('{1,2} <= a ->', {1,2}.issubset(a))
print('a >= {1,2} ->', a.issuperset({1,2}))
```

```python
# practical: remove duplicates while preserving order (use sets + list)
lst = [1, 2, 2, 3, 4, 3, 5]
# simple way (order not preserved): list(set(lst))
# to preserve order:
seen = set()
unique = []
for item in lst:
    if item not in seen:
        unique.append(item)
        seen.add(item)
print('original:', lst)
print('unique:', unique)

# frozenset (immutable set)
fs = frozenset([1,2,3])
print('frozenset:', fs)
```
