# 📘 Python Tuples Practice

Tuples are ordered, immutable sequences. This notebook covers creation, unpacking, and common tuple use-cases.

## 📘 Properties of Tuples

- **Ordered**  
  → Elements maintain their insertion order.

- **Immutable**  
  → Elements cannot be modified after creation.

- **Heterogeneous**  
  → Can store elements of different data types.

- **Allow Duplicates**  
  → Same value can appear multiple times.

- **Hashable** *(if all elements are hashable)*  
  → Can be used as dictionary keys or set elements.

## 📘 Useful Tuple Methods & Functions

### 🔹 Tuple Methods
- `count(value)`  
  → Returns the number of occurrences of a specified value.

- `index(value)`  
  → Returns the index of the first occurrence of the specified value.

---

### 🔹 Built-in Functions

- `len(tuple)`  
  → Returns the total number of elements.

- `max(tuple)`  
  → Returns the largest element.

- `min(tuple)`  
  → Returns the smallest element.

- `sum(tuple)` *(for numeric tuples)*  
  → Returns the sum of elements.

```python
# creating tuples
t1 = (1, 2, 3)
t2 = tuple([4, 5, 6])
t3 = 7, 8, 9  # packing

single = (1,)  # single-element tuple
print(single)  # prints (1,), which is a tuple

single_element = (1)  # this is just an integer, not a tuple
print(single_element)  # prints 1, not a tuple


print(t1, t2, t3, single)

# unpacking
a, b, c = t1
print(a, b, c)

# swap variables using tuple unpacking
x, y = 1, 2 #temporary tuple (1, 2) is created and unpacked into x and y
print(type(x))
print(y)
x, y = y, x #actually creates a temporary tuple (y, x) and unpacks it back into x and y
print('swapped:', x, y)
```

```python
# immutability demonstration
t = (1, 2, [3, 4])
#t[0] = 10  # TypeError if uncommented
# but nested mutable objects can be modified
t[2].append(5)
print(t)

# tuple as dict key (if elements are hashable)
#dictionary allows only immutable types as keys, and tuples can be used as keys 
#if all their elements are hashable (like numbers, strings, etc.)
d = { (1,2): 'a', (3,4): 'b' }
print(d[(1,2)])

# methods
t4 = (1,2,2,3)
print('count 2 ->', t4.count(2))
print('index of 3 ->', t4.index(3))
```

```python
# converting between list and tuple
lst = [1,2,3]
tpl = tuple(lst)
lst2 = list(tpl)
print(tpl, lst2)

# iteration and slicing work similarly to lists
print(tpl[0], tpl[1:])

# use-case: returning multiple values from a function
def stats(nums):
    return min(nums), max(nums), sum(nums)
mn, mx, s = stats([1,2,3,4])
print(mn, mx, s)
```

## ➕ Additional Tuple Topics

- **Extended unpacking** using `*` to capture multiple values.
- **Concatenation & repetition** with `+` and `*`.
- **Constructing a tuple from a generator** using `tuple(...)`.
- **`collections.namedtuple`** for lightweight, readable records.

```python
# extended unpacking
t = (1, 2, 3, 4, 5)
a, *middle, e = t
print('a:', a, 'middle:', middle, 'e:', e)

# concatenation and repetition
t1 = (1, 2)
t2 = (3, 4)
print('concat:', t1 + t2)
print('repeat:', t1 * 3)

# tuple from generator expression
tpl = tuple(x*x for x in range(5))
print('squares tuple:', tpl)

# namedtuple example
from collections import namedtuple
Point = namedtuple('Point', 'x y')
p = Point(10, 20)
print('point:', p, 'x:', p.x, 'y:', p.y)
```

## ✅ More Common Tuple Operations

- **Membership test** with `in` to check presence of an element.
- **Lexicographic comparison** between tuples (uses element-wise comparison).
- **Packing/unpacking in function calls** using `*` and `*args`/`**kwargs` patterns.
- **Copying a tuple**: use `tuple(...)` or slicing; tuples have no `.copy()` method because they are immutable.

```python
# membership and comparison
t = (1, 2, 3)
print(2 in t, 5 in t)  # True False
print((1, 2) < (1, 3))  # lexicographic comparison -> True
print((1,2,3) == (1,2,3))

# packing/unpacking in function calls
def add(a, b, c):
    return a + b + c
vals = (10, 20, 30)
print('add(*vals) ->', add(*vals))
def show(*args):
    print('received args tuple:', args)
show(*vals)

# copying a tuple (no .copy() method)
t_copy = tuple(t)
print('copied via tuple():', t_copy)
try:
    t.copy()
except Exception as e:
    print('tuples have no .copy():', type(e).__name__)
```
