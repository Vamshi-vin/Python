```python
#Quick Sort implementation
#explaination: Quick Sort is a divide-and-conquer algorithm that sorts an array by selecting a 'pivot' element and partitioning the other elements into three sub-arrays: those less than the pivot, those equal to the pivot, and those greater than the pivot. The sub-arrays are then sorted recursively. This implementation uses list comprehensions for partitioning and is not in-place, meaning it creates new lists for the partitions.
#complexity: The average and best-case time complexity of Quick Sort is O(n log n), 
# while the worst-case time complexity is O(n^2) 
# (which occurs when the smallest or largest element is always chosen as the pivot). 
# However, with good pivot selection strategies (like using the median), 
# the worst-case can be avoided, keeping the time complexity at O(n log n).
# good pivot selection strategies include:
# 1. Random Pivot: Randomly select a pivot element from the array.
# 2. Median Pivot: Use the median of the first, middle, and last elements as the pivot. 


def quick_sort(arr):
    if len(arr) <= 1:
        return arr
    pivot = arr[len(arr) // 2]  # Choosing the middle element as pivot
    left = [x for x in arr if x < pivot]  # Elements less than pivot
    middle = [x for x in arr if x == pivot]  # Elements equal to pivot
    right = [x for x in arr if x > pivot]  # Elements greater than pivot
    return quick_sort(left) + middle + quick_sort(right)  # Recursively sort and combine    

# Example usage
arr = [3, 6, 8, 10, 1, 2, 1]
sorted_arr = quick_sort(arr)
print(f"Sorted array: {sorted_arr}")  # Output: [1, 1, 2, 3, 6, 8, 10]
```

```python
#bubble sort implementation
#explaination: Bubble Sort is a simple sorting algorithm that repeatedly steps through the list,
# compares adjacent elements and swaps them if they are in the wrong order. The process is repeated until the list is sorted. This implementation includes an optimization to stop the algorithm if no swaps were made, indicating that the list is already sorted.
#complexity: The time complexity of Bubble Sort is O(n^2) in the worst and average cases, and O(n) in the best case (when the array is already sorted). 
def bubble_sort(arr):
    n = len(arr)
    for i in range(n):
        swapped = False  # Flag to check if a swap occurred
        for j in range(0, n-i-1):
            if arr[j] > arr[j+1]:  # Compare adjacent elements
                arr[j], arr[j+1] = arr[j+1], arr[j]  # Swap if they are in the wrong order
                swapped = True  # Set flag to True if a swap occurred
        if not swapped:  # If no swaps were made, the array is already sorted
            break   
# Example usage
arr = [64, 34, 25, 12, 22, 11, 90]
bubble_sort(arr)
print(f"Sorted array: {arr}")  # Output: [11, 12, 22, 25, 34, 64, 90]
```

```python
#merge sort implementation
#explaination: Merge Sort is a divide-and-conquer algorithm that divides the input array
# into two halves, calls itself for the two halves, and then merges the two sorted halves. 
# The merge() function is used for merging two halves. 
# The merge(arr, l, m, r) is a key process that assumes that arr[l..m] and arr[m+1..r] are sorted 
# and merges the two sorted sub-arrays into one.

#Explain the flow for merge sort:
#1. The merge_sort function is called with the input array.
#2. The function checks if the length of the array is greater than 1. If it is, it proceeds to divide the array into two halves.
#3. The mid index is calculated, and the array is sliced into two sub-arrays: L (left half) and R (right half).
#4. The merge_sort function is called recursively on both sub-arrays (L and R) to sort them.
#5. After the recursive calls, the function initializes three pointers: i, j, and k, which are used to track the current index of the left half, right half, and the merged array, respectively.
#6. The function enters a while loop that continues until either the left half (L) or the right half (R) is fully traversed. Inside the loop, it compares the current elements of L and R. The smaller element is added to the merged array (arr), and the corresponding pointer (i or j) is increment
#7. After the first while loop, there may be remaining elements in either L or R. The function checks for any remaining elements in L and adds them to arr, and then does the same for R.
#8. Finally, the merged array (arr) is sorted, and the function returns.

def merge_sort(arr):
    if len(arr) > 1:
        mid = len(arr) // 2  # Finding the mid of the array
        L = arr[:mid]        # Dividing the array elements into 2 halves
        R = arr[mid:]
        print(f"Dividing: {arr} into {L} and {R}")
        merge_sort(L)        # Sorting the first half
        merge_sort(R)        # Sorting the second half

        i = j = k = 0
        
        # Copy data to temp arrays L[] and R[]
        while i < len(L) and j < len(R):
            print(f"Comparing: {L[i]} and {R[j]}")
            if L[i] < R[j]:
                arr[k] = L[i]
                i += 1
            else:
                arr[k] = R[j]
                j += 1
            k += 1
        print(f"Merging: {L} and {R} into {arr}")

        
        # Checking if any element was left
        while i < len(L):
            arr[k] = L[i]
            i += 1
            k += 1
        
        while j < len(R):
            arr[k] = R[j]
            j += 1
            k += 1
# Example usage
arr = [38, 27, 43, 3, 9, 82, 10]
merge_sort(arr)
print(f"Sorted array: {arr}")  # Output: [3, 9, 10, 27, 38, 43, 82]
```

```python
#bubble sort implementation
#explaination: Bubble Sort is a simple sorting algorithm that repeatedly steps through the list,
# compares adjacent elements and swaps them if they are in the wrong order. The process is repeated until the list is sorted. This implementation includes an optimization to stop the algorithm if no swaps were made, indicating that the list is already sorted.
#complexity: The time complexity of Bubble Sort is O(n^2) in the worst and average cases, and O(n) in the best case (when the array is already sorted). 
def bubble_sort(arr):
    n = len(arr)
    for i in range(n):
        swapped = False  # Flag to check if a swap occurred
        for j in range(0, n-i-1):
            if arr[j] > arr[j+1]:  # Compare adjacent elements
                arr[j], arr[j+1] = arr[j+1], arr[j]  # Swap if they are in the wrong order
                swapped = True  # Set flag to True if a swap occurred
        if not swapped:  # If no swaps were made, the array is already sorted
            break  
# Example usage
arr = [64, 34, 25, 12, 22, 11, 90]
bubble_sort(arr)
print(f"Sorted array: {arr}")  # Output: [11, 12, 22, 25, 34, 64, 90]       
```

```python
#insertion sort implementation
#explaination: Insertion Sort is a simple sorting algorithm that builds the sorted array one item at a time. It works by repeatedly taking the next unsorted element and inserting it into the correct position in the already sorted part of the array. This implementation iterates through the array, comparing each element with the elements before it and shifting larger elements to the right until it finds the correct position for the current element.
#complexity: The time complexity of Insertion Sort is O(n^2) in the worst and average cases, and O(n) in the best case (when the array is already sorted). This is because in the worst case, each element needs to be compared with all the other elements before it, while in the best case, each element is already in the correct position, resulting in only one comparison per element.
def insertion_sort(arr):            
    for i in range(1, len(arr)):
        key = arr[i]  # The current element to be compared
        j = i - 1    # The index of the last sorted element
        print(f"Current key: {key}, comparing with sorted part: {arr[:i]}")
        while j >= 0 and key < arr[j]:  # Move elements of arr[0..i-1], that are greater than key, to one position ahead of their current position
            arr[j + 1] = arr[j]
            j -= 1
        arr[j + 1] = key  # Place the key in its correct position in the sorted part
        print(f"Array after inserting {key}: {arr}")
# Example usage
arr = [12, 11, 13, 5, 6]
insertion_sort(arr)
print(f"Sorted array: {arr}")  # Output: [5, 6, 11, 12, 13]         
```
