# 🔄 Python Loops Practice

Loops iterate over sequences and execute code blocks repeatedly. This notebook covers `for` loops, `while` loops, loop control, and advanced iteration patterns.

## 📘 Types of Loops & Loop Control

- **For loops**: Iterate over iterables (lists, tuples, strings, dicts, sets, ranges).
- **While loops**: Repeat while a condition is true.
- **Loop control**: `break` (exit loop), `continue` (skip iteration), `else` (runs if loop completes normally).
- **Nested loops**: Loops inside loops for 2D/3D structures.

## 🔹 Common Loop Functions & Patterns

- `range(start, stop, step)` — generate numbers
- `enumerate(iterable, start=0)` — pairs index and value
- `zip(iter1, iter2, ...)` — combine multiple iterables
- `reversed(iterable)` — iterate in reverse
- List/dict/set comprehensions — compact loop syntax
- Generator expressions — memory-efficient iteration

```python
# basic for loop over a list
fruits = ['apple', 'banana', 'cherry']
for fruit in fruits:
    print(fruit)

# for loop over a string
for char in 'hello':
    print(char)

# for loop with range
for i in range(3):  # 0, 1, 2
    print(i)

# range with start, stop, step
for i in range(1, 6, 2):  # 1, 3, 5
    print(i)
```

```python
# while loop
count = 0
while count < 3:
    print('count:', count)
    count += 1

# while with break
i = 0
while True:
    if i == 3:
        break
    print(i)
    i += 1

# while with continue
n = 0
while n < 5:
    n += 1
    if n == 3:
        continue  # skip when n equals 3
    print(n)
```

```python
# for-else: else runs if loop completes normally (not broken)
for i in range(3):
    print(i)
else:
    print('loop completed normally')

# for-else with break (else skipped)
for i in range(5):
    if i == 2:
        print('breaking early')
        break
    print(i)
else:
    print('this will not print because we broke')
```

```python
# enumerate - get index and value
colors = ['red', 'green', 'blue']
for idx, color in enumerate(colors):
    print(f'{idx}: {color}')

# enumerate with start parameter
for idx, color in enumerate(colors, start=1):
    print(f'{idx}: {color}')
```

```python
# Parse CSV row and match values to column names
headers = ['name', 'age', 'email']
row = ['Alice', 30, 'alice@example.com']
for col_idx, col_name in enumerate(headers):
    print(f"{col_name}: {row[col_idx]}")
```

```python
# Generate numbered list items
items = ['Buy milk', 'Pay bills', 'Call mom']
for idx, item in enumerate(items, start=1):
    print(f"<li>{idx}. {item}</li>")
```

```python
# zip - combine multiple iterables
nums = [1, 2, 3]
letters = ['a', 'b', 'c']
for num, letter in zip(nums, letters):
    print(num, letter)

# zip with unequal lengths (stops at shortest)
long_list = [1, 2, 3, 4, 5]
short_list = ['x', 'y']
for item1, item2 in zip(long_list, short_list):
    print(item1, item2)

# zip unpacking multiple sequences
zipped = list(zip([1,2,3], ['a','b','c'], [True, False, True]))
print('zipped:', zipped)
```

```python
# reversed() - iterate in reverse
lst = [1, 2, 3, 4]
for item in reversed(lst):
    print(item)

# note: reversed() returns an iterator, not a list
rev = reversed(lst)
print('type:', type(rev))
print('list from reversed:', list(rev))

# slicing [::-1] vs reversed()
print('slice [::-1]:', lst[::-1])
```

```python
# nested loops - cartesian product
for i in range(2):
    for j in range(2):
        print(f'({i}, {j})')

# nested loop example: print multiplication table
print('multiplication table:')
for i in range(1, 4):
    for j in range(1, 4):
        print(f'{i}*{j}={i*j}', end='  ')
    print()  # newline
```

```python
# list comprehension - compact loop syntax
squares = [x*x for x in range(5)]
print('squares:', squares)

# list comprehension with condition
evens = [x for x in range(10) if x % 2 == 0]
print('evens:', evens)

# nested list comprehension
matrix = [[i*j for j in range(3)] for i in range(3)]
print('matrix:', matrix)

# dict comprehension
d = {x: x*x for x in range(5)}
print('dict:', d)

# set comprehension
s = {x % 3 for x in range(10)}  # unique remainders when dividing by 3
print('set:', s)
```

```python
# generator expression - like list comprehension but memory-efficient
gen = (x*x for x in range(5))  # parentheses, not brackets
print('generator:', gen)
print('type:', type(gen))

# generator is lazy - values computed on demand
print('next value:', next(gen))
print('next value:', next(gen))

# convert to list to see all values
gen2 = (x*x for x in range(5))
print('all values:', list(gen2))
```

```python
# iterating over dictionaries
person = {'name': 'Alice', 'age': 30, 'city': 'NYC'}

# iterate keys (default)
for key in person:
    print('key:', key)

# iterate values
for value in person.values():
    print('value:', value)

# iterate items (key-value pairs)
for key, value in person.items():
    print(f'{key} -> {value}')
```

## ✅ Advanced Loop Topics

- **`itertools` module** for advanced iteration patterns.
- **Custom iterators** using `__iter__()` and `__next__()`.
- **`any()` / `all()`** to short-circuit loops on boolean sequences.

```python
# any() - returns True if any element is truthy
print(any([False, False, True]))  # True
print(any([False, False, False]))  # False

# all() - returns True if all elements are truthy
print(all([True, True, True]))  # True
print(all([True, False, True]))  # False

# practical: check if any value is negative
nums = [1, -2, 3]
if any(n < 0 for n in nums):
    print('contains negative')
```

```python
# itertools examples
from itertools import count, cycle, repeat, chain, combinations, permutations

# count() - infinite counter
counter = count(10, 2)  # start=10, step=2
for _ in range(5):
    print(next(counter), end=' ')
print()

# cycle() - repeat iterable indefinitely
colors_cycle = cycle(['red', 'green', 'blue'])
for _ in range(6):
    print(next(colors_cycle), end=' ')
print()

# repeat() - repeat value n times
repeated = repeat('x', 3)
print('repeat:', list(repeated))

# chain() - concatenate iterables
chained = chain([1,2], [3,4], ['a','b'])
print('chain:', list(chained))
```

```python
# combinations and permutations
from itertools import combinations, permutations

# combinations - unordered selections
combo = combinations([1,2,3], 2)
print('combinations:', list(combo))

# permutations - ordered selections
perm = permutations([1,2,3], 2)
print('permutations:', list(perm))
```

**Key takeaways:** Use `for` for iterables, `while` for conditions, comprehensions for compact code, generators for efficiency, and `itertools` for advanced patterns.
