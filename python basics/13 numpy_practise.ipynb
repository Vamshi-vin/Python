# 🧮 NumPy Interview Preparation

This notebook collects high-value NumPy topics and concise examples you’ll commonly see in interviews: array creation, indexing, broadcasting, vectorized operations, performance, linear algebra, random, and common pitfalls.

## Setup: import NumPy and check version

```python
import numpy as np
print('NumPy version:', np.__version__)
```

## Array creation (common constructors)

```python
# from list
a = np.array([1, 2, 3])
# zeros, ones, full, arange, linspace, eye, empty
z = np.zeros((2,3))
o = np.ones(3)
f = np.full((2,2), 7)
r = np.arange(0, 10, 2)
l = np.linspace(0, 1, 5)
I = np.eye(3)
print(a, ' \n', z, '\n ', o, ' \n', f, '\n ', r, '\n ', l, ' \n', I)
```

## Shapes, dtype, and basic attributes

```python
arr = np.array([[1,2,3],[4,5,6]])
print('ndim:', arr.ndim)
print('shape:', arr.shape)
print('size:', arr.size)
print('dtype:', arr.dtype)
```

## Indexing, slicing, fancy indexing and boolean masks

```python
x = np.arange(10)
print(x[2:6])  # slice
print(x[[1,3,5]])  # fancy indexing (creates copy)
mask = x % 2 == 0
print('even:', x[mask])  # boolean indexing
# 2D indexing
m = np.arange(12).reshape(3,4)
print(m[1,2], m[0,:], m[:,1])
```

## Views vs Copies (very important interview topic)

```python
a = np.arange(10)
s = a[2:5]  # view (usually)
s[0] = 99
print('a after modifying view:', a)
# fancy indexing returns copy (no change to original)
b = a[[2,3,4]]
b[0] = -1
print('a after fancy-index copy modified:', a)
```

## Broadcasting rules (concise explanation)

```python
# broadcasting example: add vector to each row of matrix
mat = np.array([[1,2,3],[4,5,6]])
vec = np.array([10, 20, 30])
print(mat + vec)
# broadcasting with different shapes: (3,1) with (3,)
col = np.array([[1],[2],[3]])
row = np.array([10,20,30])
print(col + row)
```

## Vectorized operations and reductions (fast vs Python loops)

```python
a = np.arange(1_000_000)
# vectorized sum (C-optimized)
%timeit a.sum()
# Python loop (much slower)
def py_sum(x):
    s = 0
    for v in x:
        s += v
    return s

%timeit py_sum(a)
```

## Reshape, transpose, ravel, flatten, stacking

```python
x = np.arange(12)
print('reshape:', x.reshape(3,4))
print('transpose:', x.reshape(3,4).T)
print('ravel (view):', x.reshape(3,4).ravel())
print('flatten (copy):', x.reshape(3,4).flatten())
# stacking
a = np.array([1,2,3])
b = np.array([4,5,6])
print('vstack:', np.vstack([a,b]))
print('hstack:', np.hstack([a,b]))
```

## Datatypes and casting (astype)

```python
x = np.array([1.5, 2.7, 3.1])
print(x.dtype)
y = x.astype(int)
print(y, y.dtype)
# careful with overflow and precision when casting
```

## Linear algebra (dot, matmul, inv, svd)

```python
A = np.array([[1.,2.],[3.,4.]])
b = np.array([1., 2.])
print('dot:', A.dot(b))
print('matmul:', A @ b)
print('inv:', np.linalg.inv(A))
print('eig:', np.linalg.eig(A))
print('svd:', np.linalg.svd(A))
```

## Random module (np.random) and seeding for reproducibility

```python
np.random.seed(0)
print(np.random.rand(3))
print(np.random.randint(0, 10, size=(2,3)))
```

## Performance tips and memory layout (C vs F order)

```python
x = np.arange(12).reshape(3,4)
print('C contiguous:', x.flags['C_CONTIGUOUS'])
print('F contiguous:', x.flags['F_CONTIGUOUS'])
xf = np.asfortranarray(x)
print('after asfortran C contiguous:', xf.flags['C_CONTIGUOUS'])
print('after asfortran F contiguous:', xf.flags['F_CONTIGUOUS'])
```

## Common interview questions / exercises (with hints)

1. Given a 1D array, return the indices of the top-k largest elements.
2. Given a 2D array, compute the row-wise mean without explicit Python loops.
3. Implement a vectorized sliding window sum (convolution) for 1D arrays.
4. Given boolean mask and array, set all masked positions to 0 efficiently.
5. Explain broadcasting rules with examples.

```python
# Example solution: top-k indices (question 1)
arr = np.random.rand(10)
k = 3
idx = np.argpartition(-arr, k-1)[:k]  # O(n) selection
print('top-k values:', arr[idx])
# sorted order of top-k
print('sorted top-k:', arr[np.argsort(-arr[idx])])
```

## Gotchas and pitfalls

- Be careful with shapes when broadcasting (read error messages).
- Watch out for integer division vs float (use dtype=float when needed).
- In-place operations can change dtypes or views unexpectedly.
- Use `np.copy()` when you need a true copy.
- Pay attention to memory order when interfacing with C/Fortran libraries.
