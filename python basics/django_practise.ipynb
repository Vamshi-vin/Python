# Django Web Framework - Comprehensive Guide

Django is a high-level Python web framework that encourages rapid development and clean pragmatic design. It follows the **Model-View-Controller (MVC)** architecture pattern (called **MVT** in Django - Model-View-Template).

## What is Django?

Django is a full-featured web framework that includes:
- **ORM (Object-Relational Mapping)** for database interaction
- **Admin Panel** for data management
- **Authentication & Authorization** built-in
- **URL Routing** system
- **Template Engine** for HTML rendering
- **Form Handling** and validation
- **Security Features** (CSRF, XSS protection)
- **Middleware** support

- **templates/**: HTML files for rendering

## 3. Django MVT Architecture

Django uses the **Model-View-Template (MVT)** pattern:

```
┌─────────────────────────────────────────────────────────┐
│                    HTTP Request                          │
└─────────────────────────────────────────────────────────┘
                           ↓
┌─────────────────────────────────────────────────────────┐
│                    URL Router (urls.py)                  │
│              Matches URL pattern and calls View          │
└─────────────────────────────────────────────────────────┘
                           ↓
┌─────────────────────────────────────────────────────────┐
│                   View (views.py)                        │
│        Processes request, queries Model, passes          │
│              data to Template                            │
└─────────────────────────────────────────────────────────┘
                           ↓
                    ┌──────┴──────┐
                    ↓             ↓
            ┌─────────────┐  ┌──────────┐
            │  Model      │  │ Template │
            │ (models.py) │  │(HTML)    │
            └─────────────┘  └──────────┘
                    ↓             ↓
                    └──────┬──────┘
                           ↓
┌─────────────────────────────────────────────────────────┐
│                  HTTP Response                           │
│              (Rendered HTML to browser)                  │
└─────────────────────────────────────────────────────────┘
```

### Components:

| Component | Purpose |
|-----------|---------|
| **Model** | Database layer - defines data structure |
| **View** | Business logic - processes requests |
| **Template** | Presentation layer - renders HTML |

## 6. Django Views - Business Logic Layer

Views handle the request-response cycle. There are two types: **Function-Based Views (FBV)** and **Class-Based Views (CBV)**.

````python
# Django Views Examples
views_code = """
from django.shortcuts import render, get_object_or_404, redirect
from django.http import HttpResponse, JsonResponse
from django.views import View
from django.views.generic import ListView, DetailView, CreateView, UpdateView, DeleteView
from django.contrib.auth.mixins import LoginRequiredMixin
from .models import Book, Author
from .forms import BookForm

# ============== Function-Based Views (FBV) ==============

# Simple view
def home(request):
    return HttpResponse("Welcome to Django!")

# View with template
def book_list(request):
    books = Book.objects.all()
    context = {'books': books}
    return render(request, 'books/book_list.html', context)

# View with filtering
def author_books(request, author_id):
    author = get_object_or_404(Author, id=author_id)
    books = author.book_set.all()
    return render(request, 'books/author_books.html', {'author': author, 'books': books})

# View with form handling
def create_book(request):
    if request.method == 'POST':
        form = BookForm(request.POST)
        if form.is_valid():
            book = form.save()
            return redirect('book_detail', pk=book.id)
    else:
        form = BookForm()
    return render(request, 'books/create_book.html', {'form': form})

# JSON response
def api_books(request):
    books = Book.objects.values('id', 'title', 'price')
    return JsonResponse(list(books), safe=False)

# ============== Class-Based Views (CBV) ==============

# List view
class BookListView(ListView):
    model = Book
    template_name = 'books/book_list.html'
    context_object_name = 'books'
    paginate_by = 10
    
    def get_queryset(self):
        return Book.objects.filter(is_available=True)

# Detail view
class BookDetailView(DetailView):
    model = Book
    template_name = 'books/book_detail.html'
    context_object_name = 'book'

# Create view
class BookCreateView(CreateView):
    model = Book
    form_class = BookForm
    template_name = 'books/book_form.html'
    success_url = '/books/'

# Update view
class BookUpdateView(UpdateView):
    model = Book
    form_class = BookForm
    template_name = 'books/book_form.html'
    success_url = '/books/'

# Delete view
class BookDeleteView(DeleteView):
    model = Book
    template_name = 'books/book_confirm_delete.html'
    success_url = '/books/'

# Protected view (requires login)
class ProtectedBookView(LoginRequiredMixin, ListView):
    model = Book
    template_name = 'books/protected_books.html'
    login_url = 'login'

# Custom class-based view
class CustomAuthorsView(View):
    def get(self, request):
        authors = Author.objects.all()
        return render(request, 'books/authors.html', {'authors': authors})
    
    def post(self, request):
        name = request.POST.get('name')
        author = Author.objects.create(name=name)
        return redirect('author_detail', pk=author.id)
"""

print(views_code)
````

````python
# Django URL Routing Examples
urls_code = """
# project/urls.py (Main URL configuration)
from django.contrib import admin
from django.urls import path, include

urlpatterns = [
    path('admin/', admin.site.urls),
    path('books/', include('books.urls')),      # Include app URLs
    path('accounts/', include('accounts.urls')),
]

# books/urls.py (App-specific URLs)
from django.urls import path
from . import views

app_name = 'books'

urlpatterns = [
    # Function-based views
    path('', views.book_list, name='book_list'),
    path('<int:book_id>/', views.book_detail, name='book_detail'),
    path('author/<int:author_id>/', views.author_books, name='author_books'),
    path('create/', views.create_book, name='create_book'),
    
    # Class-based views
    path('cbv/', views.BookListView.as_view(), name='book_list_cbv'),
    path('cbv/<int:pk>/', views.BookDetailView.as_view(), name='book_detail_cbv'),
    path('cbv/create/', views.BookCreateView.as_view(), name='book_create_cbv'),
    path('cbv/<int:pk>/update/', views.BookUpdateView.as_view(), name='book_update_cbv'),
    path('cbv/<int:pk>/delete/', views.BookDeleteView.as_view(), name='book_delete_cbv'),
    
    # API endpoint
    path('api/', views.api_books, name='api_books'),
]

# URL Path Converters:
# <int:id>      - Integer
# <str:name>    - String (default)
# <slug:slug>   - Slug (lowercase, numbers, hyphens, underscores)
# <uuid:id>     - UUID
# <path:path>   - Path including slashes

# Examples with converters:
# path('articles/<int:year>/<int:month>/', views.article_archive)
# path('files/<path:file_path>/', views.file_download)
# path('users/<slug:username>/', views.user_profile)

# Reverse URL (in templates):
# {% url 'books:book_detail' book.id %}
# {% url 'books:book_list' %}

# Reverse URL (in views):
# from django.urls import reverse
# url = reverse('books:book_detail', args=[book.id])
"""

print(urls_code)
````

````python
# Django Template Examples
template_code = """
<!-- books/templates/books/book_list.html -->
<!DOCTYPE html>
<html>
<head>
    <title>Books List</title>
    {% load static %}
    <link rel=\"stylesheet\" href=\"{% static 'css/style.css' %}\">
</head>
<body>
    <h1>Available Books</h1>
    
    <!-- If statement -->
    {% if books %}
        <ul>
        <!-- For loop -->
        {% for book in books %}
            <li>
                <a href=\"{% url 'books:book_detail' book.id %}\">
                    {{ book.title }}
                </a>
                by {{ book.author.name }}
                - ${{ book.price }}
                
                <!-- Conditional inside loop -->
                {% if book.price > 50 %}
                    <span style=\"color: red;\">Premium</span>
                {% else %}
                    <span style=\"color: green;\">Standard</span>
                {% endif %}
            </li>
        {% empty %}
            <p>No books found</p>
        {% endfor %}
        </ul>
    {% else %}
        <p>No books available</p>
    {% endif %}
    
    <!-- Filter/Format examples -->
    <p>Date: {{ book.published_date|date:\"Y-m-d\" }}</p>
    <p>Price: {{ book.price|floatformat:2 }}</p>
    <p>Title: {{ book.title|upper }}</p>
    
    <!-- Template inheritance -->
</body>
</html>

<!-- base.html (Base template) -->
<!DOCTYPE html>
<html>
<head>
    <title>{% block title %}My Site{% endblock %}</title>
</head>
<body>
    <nav>
        <a href=\"/\">Home</a>
        <a href=\"/books/\">Books</a>
    </nav>
    
    {% block content %}
    {% endblock %}
    
    <footer>© 2024 My Library</footer>
</body>
</html>

<!-- book_detail.html (Child template) -->
{% extends \"base.html\" %}

{% block title %}{{ book.title }} - My Site{% endblock %}

{% block content %}
    <div class=\"book-detail\">
        <h1>{{ book.title }}</h1>
        <p><strong>Author:</strong> {{ book.author.name }}</p>
        <p><strong>ISBN:</strong> {{ book.isbn }}</p>
        <p><strong>Price:</strong> ${{ book.price }}</p>
        
        {% if user.is_authenticated %}
            <a href=\"{% url 'books:book_update' book.id %}\">Edit</a>
            <a href=\"{% url 'books:book_delete' book.id %}\">Delete</a>
        {% endif %}
    </div>
{% endblock %}

<!-- Common Template Tags and Filters -->
Filters: {{ variable|filter_name }}
- {{ text|lower }}              # Lowercase
- {{ text|upper }}              # Uppercase
- {{ text|title }}              # Title case
- {{ text|truncatewords:10 }}   # Truncate words
- {{ date|date:\"Y-m-d\" }}     # Format date
- {{ number|floatformat:2 }}    # Format float
- {{ list|length }}             # Get length
- {{ dict|default:\"N/A\" }}    # Default value

Tags: {% tag_name %}
- {% if condition %}...{% endif %}
- {% for item in items %}...{% endfor %}
- {% block name %}...{% endblock %}
- {% extends \"base.html\" %}
- {% include \"template.html\" %}
- {% url 'view_name' arg1 arg2 %}
- {% load static %}
- {% csrf_token %}
- {% comment %}...{% endcomment %}
"""

print(template_code)
````

## 9. Django Forms - Data Validation and Input

Forms handle data validation, error messages, and secure processing.

````python
# Django Forms Examples
forms_code = """
from django import forms
from django.core.exceptions import ValidationError
from .models import Book, Author

# ============== Model Forms ==============
# Auto-generated forms from models

class BookForm(forms.ModelForm):
    class Meta:
        model = Book
        fields = ['title', 'author', 'isbn', 'published_date', 'pages', 'price', 'categories']
        widgets = {
            'title': forms.TextInput(attrs={'class': 'form-control'}),
            'published_date': forms.DateInput(attrs={'type': 'date'}),
            'price': forms.NumberInput(attrs={'step': '0.01'}),
        }

# ============== Regular Forms ==============
# Regular forms not tied to models

class SearchForm(forms.Form):
    query = forms.CharField(
        label='Search',
        max_length=100,
        widget=forms.TextInput(attrs={'placeholder': 'Search books...'})
    )
    date_from = forms.DateField(required=False, widget=forms.DateInput(attrs={'type': 'date'}))
    date_to = forms.DateField(required=False, widget=forms.DateInput(attrs={'type': 'date'}))

# ============== Custom Validation ==============

class RegistrationForm(forms.Form):
    username = forms.CharField(max_length=30)
    email = forms.EmailField()
    password = forms.CharField(widget=forms.PasswordInput)
    password_confirm = forms.CharField(widget=forms.PasswordInput)
    
    # Field-level validation
    def clean_username(self):
        username = self.cleaned_data.get('username')
        if len(username) < 3:
            raise ValidationError('Username must be at least 3 characters')
        return username
    
    # Form-level validation (access multiple fields)
    def clean(self):
        cleaned_data = super().clean()
        password = cleaned_data.get('password')
        password_confirm = cleaned_data.get('password_confirm')
        
        if password != password_confirm:
            raise ValidationError('Passwords do not match')
        
        return cleaned_data

# ============== Form Field Types ==============

class CompleteForm(forms.Form):
    # Text fields
    name = forms.CharField(max_length=100)
    description = forms.CharField(widget=forms.Textarea)
    
    # Number fields
    age = forms.IntegerField()
    price = forms.DecimalField(max_digits=6, decimal_places=2)
    
    # Choice fields
    CATEGORY_CHOICES = [('tech', 'Technology'), ('fiction', 'Fiction')]
    category = forms.ChoiceField(choices=CATEGORY_CHOICES)
    tags = forms.MultipleChoiceField(choices=CATEGORY_CHOICES, widget=forms.CheckboxSelectMultiple)
    
    # Date and time
    date_field = forms.DateField()
    time_field = forms.TimeField()
    datetime_field = forms.DateTimeField()
    
    # Boolean
    agree_terms = forms.BooleanField()
    
    # Email and URL
    email = forms.EmailField()
    website = forms.URLField()
    
    # File fields
    document = forms.FileField()
    image = forms.ImageField()
    
    # Slider (with custom widget)
    rating = forms.IntegerField(min_value=1, max_value=5)

# ============== Using Forms in Views ==============

# Function-based view
def create_book(request):
    if request.method == 'POST':
        form = BookForm(request.POST)
        if form.is_valid():
            # Access cleaned data
            title = form.cleaned_data['title']
            author = form.cleaned_data['author']
            
            # Save to database
            book = form.save()
            return redirect('book_detail', pk=book.id)
    else:
        form = BookForm()
    return render(request, 'books/form.html', {'form': form})

# Class-based view (automatic)
from django.views.generic import CreateView

class BookCreateView(CreateView):
    model = Book
    form_class = BookForm
    template_name = 'books/form.html'
    success_url = '/books/'
"""

print(forms_code)
````
