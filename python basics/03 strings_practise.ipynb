# 🔤 Python Strings Practice

Strings are immutable sequences of characters. This notebook covers string creation, indexing, slicing, methods, formatting, manipulation, and validation.

## 📘 String Fundamentals

- **Creation**: Single quotes `''`, double quotes `""`, triple quotes `'''` or `"""`
- **Indexing**: Access individual characters by position; negative indexing from end
- **Slicing**: Extract substrings using `[start:stop:step]`
- **Immutability**: Strings cannot be modified in-place; operations return new strings
- **Concatenation**: Join strings with `+` or `f-strings`
- **Methods**: 80+ built-in string methods (case, split, strip, replace, etc.)

```python
# creating strings
s1 = 'single quotes'
s2 = "double quotes"
s3 = '''triple quotes
can span
multiple lines'''

print(s1)
print(s2)
print(s3)

# string length
print('length of s1:', len(s1))
```

```python
# indexing and slicing
text = 'Hello, Python!'
print('text:', text)
print('text[0]:', text[0])  # first character
print('text[-1]:', text[-1])  # last character
print('text[0:5]:', text[0:5])  # slice from 0 to 4
print('text[7:]:', text[7:])  # from index 7 to end
print('text[:5]:', text[:5])  # from start to index 4
print('text[::2]:', text[::2])  # every 2nd character
print('text[::-1]:', text[::-1])  # reversed
```

```python
# case conversion methods
text = 'Hello, World!'
print('upper():', text.upper())
print('lower():', text.lower())
print('capitalize():', text.capitalize())  # first char uppercase
print('title():', text.title())  # each word capitalized
print('swapcase():', text.swapcase())  # toggle case
```

```python
# searching methods
text = 'Hello, Python! Python is great!'
print('find():', text.find('Python'))  # returns index, -1 if not found
print('rfind():', text.rfind('Python'))  # find from right
print('count():', text.count('Python'))  # count occurrences
print('startswith():', text.startswith('Hello'))  # True/False
print('endswith():', text.endswith('!'))  # True/False
print('in operator:', 'Python' in text)  # membership test
```

```python
# strip and split methods
text = '  Hello, World!  '
print('original:', repr(text))  # repr shows whitespace
print('strip():', repr(text.strip()))  # remove from both ends
print('lstrip():', repr(text.lstrip()))  # remove from left
print('rstrip():', repr(text.rstrip()))  # remove from right

# split and join
text2 = 'apple,banana,cherry'
items = text2.split(',')  # split by delimiter
print('split():', items)
print('join():', '-'.join(items))  # join with delimiter
```

```python
# replace method
text = 'apple apple apple'
print('original:', text)
print('replace():', text.replace('apple', 'orange'))  # replace all
print('replace(count=2):', text.replace('apple', 'orange', 2))  # replace first 2

# note: strings are immutable
text = 'Hello'
# text[0] = 'J'  # TypeError if uncommented
text = 'J' + text[1:]  # create new string instead
print('after replacement:', text)
```

```python
# string formatting: f-strings (Python 3.6+) - preferred
name = 'Alice'
age = 30
print(f'Name: {name}, Age: {age}')

# expression in f-string
print(f'Next year: {age + 1}')

# formatting options
price = 19.99
print(f'Price: ${price:.2f}')  # 2 decimal places
print(f'Number: {42:05d}')  # pad with zeros
print(f'Text: {name:>10}')  # right align
print(f'Text: {name:<10}X')  # left align
```

```python
# format() method
print('Format method: {} is {}'.format(name, age))
print('{1} is {0}'.format(age, name))  # access by index
print('{n} is {a}'.format(n=name, a=age))  # named arguments

# % formatting (old style, still used sometimes)
print('Percent formatting: %s is %d' % (name, age))
```

```python
# escape sequences
print('Line1\nLine2')  # newline
print('Tab\there')  # tab
print('Backslash: \\')  # backslash
print('Quote: \'single\'')  # single quote
print('Double: "double"')  # double quote
print('Unicode: \u0041')  # Unicode character (A)

# raw string (r prefix) - backslash not escaped
print('Raw string: r"C:\\Users\\alice"')
path = r'C:\Users\alice\file.txt'
print(path)
```

```python
# isX() checking methods
text1 = '123'
text2 = 'abc'
text3 = 'Hello123'

print('isdigit():', text1.isdigit())  # all digits?
print('isalpha():', text2.isalpha())  # all letters?
print('isalnum():', text3.isalnum())  # letters and digits?
print('isspace():', '   '.isspace())  # all whitespace?
print('isupper():', 'HELLO'.isupper())  # all uppercase?
print('islower():', 'hello'.islower())  # all lowercase?
```

```python
# string operations
s1 = 'Hello'
s2 = 'World'
print('concatenation:', s1 + ' ' + s2)
print('repetition:', s1 * 3)
print('membership:', 'H' in s1, 'x' in s1)

# string comparison
print('\nComparison:')
print('s1 == "Hello":', s1 == 'Hello')
print('s1 < s2:', s1 < s2)  # lexicographic
print('"apple" < "banana":', 'apple' < 'banana')
```

```python
# converting strings
num_str = '123'
num_int = int(num_str)
num_float = float(num_str)
print(f'int: {num_int}, float: {num_float}')

# string to list
text = 'abc'
char_list = list(text)
print('list:', char_list)

# other conversions
num = 42
text = str(num)
print(f'str(42): {text}, type: {type(text)}')
```

```python
# regular expressions (basic intro)
import re

text = 'Email: alice@example.com, bob@test.org'
# find all email patterns
emails = re.findall(r'\b[\w.-]+@[\w.-]+\.\w+\b', text)
print('Emails found:', emails)

# replace with regex
masked = re.sub(r'\d', 'X', 'Phone: 555-123-4567')
print('Masked:', masked)

# check if matches
pattern = r'^\d{3}-\d{3}-\d{4}$'
print('Matches pattern:', bool(re.match(pattern, '555-123-4567')))
```

```python
# string padding and alignment
text = 'Python'
print('ljust(10):', repr(text.ljust(10)))  # left justified
print('rjust(10):', repr(text.rjust(10)))  # right justified
print('center(10):', repr(text.center(10)))  # centered

# pad with character
print('ljust(10, "-"):', text.ljust(10, '-'))
print('rjust(10, "*"):', text.rjust(10, '*'))
print('center(10, "="):', text.center(10, '='))
```

```python
# practical: validation functions
def is_valid_email(email: str) -> bool:
    """Check if email has basic valid format."""
    return '@' in email and '.' in email.split('@')[1]

def is_strong_password(pwd: str) -> bool:
    """Check if password is strong.
    At least 8 chars, 1 uppercase, 1 digit, 1 special char.
    """
    import re
    if len(pwd) < 8:
        return False
    has_upper = any(c.isupper() for c in pwd)
    has_digit = any(c.isdigit() for c in pwd)
    has_special = re.search(r'[!@#$%^&*(),.?":{}|<>]', pwd)
    return has_upper and has_digit and bool(has_special)

print('Email validation:')
print('  user@example.com:', is_valid_email('user@example.com'))
print('  invalid.email:', is_valid_email('invalid.email'))

print('\nPassword strength:')
print('  weak123:', is_strong_password('weak123'))
print('  Strong@123!:', is_strong_password('Strong@123!'))
```

## ✅ String Best Practices

- Use **f-strings** for formatting (readable and fast).
- Remember strings are **immutable**; operations return new strings.
- Use `strip()` to clean whitespace from user input.
- Use `split()` and `join()` for text processing.
- Use raw strings (`r''`) for file paths and regex patterns.
- Validate user input with `isX()` methods or regex.
- Use `startswith()` and `endswith()` instead of slicing for checks.
- Leverage regex for complex pattern matching.

```python
# practical: text processing pipeline
def clean_and_process(raw_text: str) -> list:
    """Clean text, split into words, and filter short words."""
    # remove extra whitespace
    text = raw_text.strip()
    # convert to lowercase
    text = text.lower()
    # split into words
    words = text.split()
    # filter words shorter than 3 chars
    words = [w for w in words if len(w) >= 3]
    return words

raw = '  Hello   WORLD  from   Python!  '
result = clean_and_process(raw)
print('Original:', repr(raw))
print('Processed:', result)
```
