# 📙 Python Dictionaries Practice

Dictionaries are insertion-ordered, mutable mappings from keys to values.

**Properties of dictionaries:**

- **Mapping**: Store key -> value pairs.
- **Mutable**: Add, remove, or change entries.
- **Keys must be hashable**: e.g., strings, numbers, tuples (if their elements are hashable).
- **Values can be any object**: including lists or other dicts.
- **Iteration**: iterate keys by default; use `.items()` for key/value pairs.

**Common dict methods & operations**

- `dict()` constructor, `{}` literal
- Access: `d[key]` (KeyError if missing), `d.get(key, default)`
- Insert/Update: `d[key] = value`, `d.update(other)`
- Remove: `del d[key]`, `d.pop(key, default)`, `d.popitem()`
- Views: `d.keys()`, `d.values()`, `d.items()`
- Utilities: `dict.fromkeys()`, `setdefault()`
- Comprehension: `{k: v for k, v in iterable}`

```python
# creating dictionaries
d1 = {'a': 1, 'b': 2}
d2 = dict(x=10, y=20) # using keyword arguments to create a dictionary
d3 = dict([('k1', 100), ('k2', 200)]) # using a list of tuples to create a dictionary, where each tuple is a key-value pair
print(d1, d2, d3)

# accessing values
print(d1['a'])
print(d1.get('c'))  # None if missing
print(d1.get('c', 'default'))
# print(d1['c'])  # would raise KeyError if uncommented
```

```python
# insert/update and remove
d = {'name': 'Alice', 'age': 30}
d['age'] = 31  # update
d['city'] = 'NYC'  # insert
print(d)

# remove safely
age = d.pop('age')
print('popped age:', age)
print(d)
d.pop('missing', None)  # won't raise
# del d['missing']  # would raise KeyError if uncommented
```

```python
# iteration patterns
person = {'name': 'Bob', 'skills': ['py', 'sql'], 'years': 5}
# keys
for k in person:
    print('key:', k)
# values
for v in person.values():
    print('value:', v)
# items
for k, v in person.items():
    print(k, '->', v)

# dictionary comprehension example: square mapping
squares = {i: i*i for i in range(6)}
print(squares)
```

```python
# merging dictionaries (Python 3.9+ uses | operator)
a = {'x': 1, 'y': 2}
b = {'y': 3, 'z': 4}
merged = a | b  # b wins on conflicts
print('merged:', merged)
# older pattern: {**a, **b}
print('double-star merge:', {**a, **b})

# common pattern: group items (frequency) using dict and setdefault
items = ['apple', 'banana', 'apple', 'orange', 'banana']
groups = {}
for it in items:
    groups.setdefault(it, 0)
    groups[it] += 1
print('counts:', groups)

# using collections.defaultdict simplifies this
from collections import defaultdict
dd = defaultdict(int)
for it in items:
    dd[it] += 1
print('defaultdict counts:', dict(dd))
```

```python
# nested dictionaries example and accessing safely
config = {'db': {'host': 'localhost', 'port': 5432}, 'debug': True}
# safe nested access using get
host = config.get('db', {}).get('host')
print('db host:', host)

# practical: invert a mapping (values must be unique or use lists)
orig = {'a': 1, 'b': 2, 'c': 1}
inv = {}
for k, v in orig.items():
    inv.setdefault(v, []).append(k)
print('inverted grouping:', inv)
```

---
**Quick built-ins & tips:** `len()`, `any()`, `all()`, iterate `.items()` for key/value pairs, prefer `.get()` for optional keys, use `defaultdict` for counting/ grouping.
