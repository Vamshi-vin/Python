```python
# Fastapi concepts  
# what is fastapi?
# FastAPI is a modern, fast (high-performance), web framework for building APIs with Python 3.6+ based on standard Python type hints. It is designed to be easy to use and to provide high performance, making it a popular choice for building web applications and APIs. FastAPI
# is built on top of Starlette for the web parts and Pydantic for the data handling, which allows it to provide features like automatic validation, serialization, and documentation generation. It is also designed to be easy to deploy and can be used with a variety of deployment options, including Docker and serverless platforms.

#example of a simple FastAPI application:   
from fastapi import FastAPI
app = FastAPI()
@app.get("/")
def read_root():
    return {"Hello": "World"}
# In this example, we create a FastAPI application and define a single route ("/") that returns a JSON response with the message "Hello World". To run this application, you would typically use a command like `uvicorn main:app --reload`, where `main` is the name of the Python file containing the code and `app` is the name of the FastAPI instance. 

#example of a more complex FastAPI application with path parameters and query parameters:
from fastapi import FastAPI
app = FastAPI() 
@app.get("/items/{item_id}")
def read_item(item_id: int, q: str = None):
    return {"item_id": item_id, "q": q}
# In this example, we define a route that includes a path parameter (`item_id`) and a query parameter (`q`). The `item_id` is required and must be an integer, while the `q` parameter is optional and can be a string. When you access this route, you can provide values for both parameters, and the response will include the values you provided. For example,
# if you access `/items/42?q=hello`, the response will be `{"item_id": 42, "q": "hello"}`. If you access `/items/42` without providing a query parameter, the response will be `{"item_id": 42, "q": null}`.

#example usage

```
