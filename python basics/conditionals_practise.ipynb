# Conditional Statements in Python — Practice

Examples and notes covering `if`/`elif`/`else`, truthy/falsy values, short-circuiting, ternary expressions, the walrus operator, `match`/`case` (pattern matching), and common pitfalls.

## Basic if / elif / else

```python
x = 10
if x > 0:
    print('positive')
elif x == 0:
    print('zero')
else:
    print('negative')
```

## Truthy and falsy values

```python
tests = [0, 1, '', 'hi', [], [1], {}, {'a':1}, None, False, True]
for v in tests:
    print(repr(v), '->', bool(v))
```

## Logical operators and short-circuiting

```python
def left():
    print('left called')
    return True
def right():
    print('right called')
    return False
print('and short-circuit:')
print(left() and right())  # right() *not* called when left() is False
print('or short-circuit:')
print(left() or right())  # right() not called when left() is True
```

## Ternary (conditional) expression

```python
x = 5
result = 'even' if x % 2 == 0 else 'odd'
print(result)
# chain ternaries carefully (avoid unreadable expressions)
```

## The walrus operator (`:=`) — assign inside expressions

```python
# Python 3.8+: assign and test in one expression
if (n := len([1,2,3])) > 2:
    print('length', n)
```

## `match` / `case` pattern matching (Python 3.10+) — useful for structured conditions

```python
def http_status(code):
    match code:
        case 200:
            return 'OK'
        case 404:
            return 'Not Found'
        case 500|501|502:
            return 'Server error'
        case _:
            return 'Other'

print(http_status(200), http_status(404), http_status(501), http_status(123))
```

## Common pitfalls and tips

- Avoid writing conditions that are hard to read; prefer clarity over cleverness.
- Beware of chained comparisons (they are powerful and safe: `a < b < c`).
- Do not use mutable defaults in functions for conditional logic side-effects.
- When testing membership, prefer `in`/`not in` for collections.
- Use explicit boolean checks when `None` is a valid value (avoid `if x:` if `0` or `''` are meaningful).
