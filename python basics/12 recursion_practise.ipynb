## Recursion basics and base case

```python
# factorial: classic example
def factorial(n):
    if n <= 1:
        return 1
    print(f'Calculating factorial({n}), need factorial({n-1})')
    return n * factorial(n-1)

print(factorial(5))  # 120
```

## Recursive vs iterative and recursion depth

```python
# Fibonacci: naive recursive (exponential) vs iterative (linear)
def fib_rec(n):
    if n < 2:
        return n
    return fib_rec(n-1) + fib_rec(n-2)

def fib_iter(n):
    a, b = 0, 1
    for _ in range(n):
        a, b = b, a + b
    return a

print('fib_rec(10)=', fib_rec(10))
print('fib_iter(10)=', fib_iter(10))

import sys
print('recursion limit:', sys.getrecursionlimit())
# caution: increasing recursion limit can cause C stack overflow
```

## Memoization and caching (make recursion practical)

```python
from functools import lru_cache
@lru_cache(maxsize=None)
def fib_memo(n):
    if n < 2:
        return n
    return fib_memo(n-1) + fib_memo(n-2)

print('fib_memo(35)=', fib_memo(35))  # fast compared to naive recursion
```

## Tail recursion note (Python has no TCO) — write iterative loops for deep recursion

```python
def fact_tail(n, acc=1):
    if n == 0:
        return acc
    return fact_tail(n-1, acc*n)

print(fact_tail(5))  # correct, but Python won't optimize away frames
```

## Divide-and-conquer examples: merge sort and quicksort

```python
def merge_sort(arr):
    if len(arr) <= 1:
        return arr
    mid = len(arr) // 2
    left = merge_sort(arr[:mid])
    right = merge_sort(arr[mid:])
    i = j = 0
    out = []
    while i < len(left) and j < len(right):
        if left[i] <= right[j]:
            out.append(left[i]); i += 1
        else:
            out.append(right[j]); j += 1
    out.extend(left[i:]); out.extend(right[j:])
    return out

def quicksort(arr):
    if len(arr) <= 1:
        return arr
    pivot = arr[len(arr)//2]
    left = [x for x in arr if x < pivot]
    middle = [x for x in arr if x == pivot]
    right = [x for x in arr if x > pivot]
    return quicksort(left) + middle + quicksort(right)

print('merge_sort ->', merge_sort([3,1,4,1,5,9,2]))
print('quicksort ->', quicksort([3,1,4,1,5,9,2]))
```

## Backtracking examples: permutations, combinations, N-Queens (short)

```python
# generate subsets (power set)
def subsets(nums):
    res = []
    def backtrack(i, path):
        if i == len(nums):
            res.append(path[:]); return
        # exclude
        backtrack(i+1, path)
        # include
        path.append(nums[i])
        backtrack(i+1, path)
        path.pop()
    backtrack(0, [])
    return res

print('subsets([1,2,3]) ->', subsets([1,2,3]))

# permutations (backtracking)
def permutations(nums):
    res = []
    used = [False]*len(nums)
    def backtrack(path):
        if len(path) == len(nums):
            res.append(path[:]); return
        for i in range(len(nums)):
            if used[i]: continue
            used[i] = True
            path.append(nums[i])
            backtrack(path)
            path.pop(); used[i] = False
    backtrack([])
    return res

print('permutations([1,2,3]) ->', permutations([1,2,3]))
```

## Recursion on trees (binary tree traversal example)

```python
from collections import namedtuple
Node = namedtuple('Node', ['val','left','right'])
root = Node(1, Node(2, Node(4,None,None), Node(5,None,None)), Node(3,None,Node(6,None,None)))

def inorder(node):
    if node is None: return []
    return inorder(node.left) + [node.val] + inorder(node.right)

print('inorder ->', inorder(root))
```

## Converting recursion to iterative approach (stack) — when needed

```python
# iterative inorder traversal using explicit stack
def inorder_iter(node):
    stack = []
    res = []
    cur = node
    while cur or stack:
        while cur:
            stack.append(cur); cur = cur.left
        cur = stack.pop()
        res.append(cur.val)
        cur = cur.right
    return res

print('inorder_iter ->', inorder_iter(root))
```

## Common pitfalls, performance notes and exercises

- Python recursion depth is limited; prefer iterative solutions for very deep recursions.
- Use `functools.lru_cache` to memoize overlapping subproblems (DP).
- Backtracking explores search trees; ensure pruning when possible.
- Avoid huge intermediate concatenations inside recursion (e.g., repeated `+` on lists).
